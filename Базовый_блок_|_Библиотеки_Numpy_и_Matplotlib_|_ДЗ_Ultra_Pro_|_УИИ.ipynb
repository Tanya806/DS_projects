### Задача 1

Напишите функцию матричного умножения трёх матриц без использования встроенных средств `np.dot()`, **@**.

```python
# Ваше решение
# Определим три матрицы для тестирования A, B, C
A = [[5, 4, 3], [2, 4, 6]] 
B = [[3, 2, 4, 4], [4, 3, 6, 4], [2, 7, 5,4 ]] 
C = [[3, 2, 4], [4, 3, 6], [2, 7, 5], [0,1,0]] 

#Функция перемножения двух матриц
def mult_matrix(A, B):
  multiResult = np.zeros((len(A), len(B[0])))
  if len(A[0]) != len(B):
    print("Невозможно перемножить матрицы данных размеров")
    return []
  else:
    for m in range(len(A)): 
      for n in range(len(B[0])): 
        for o in range(len(B)): 
          multiResult[m][n] += A[m][o] * B[o][n] 
    return multiResult 

#Функция перемножения трех матриц
def mult_3_matrix(A,B,C):
  res1 = mult_matrix(A, B)
  return mult_matrix(res1, C)

mt = mult_3_matrix(A, B, C)
if mt != []:
  print("Результат умножения матриц: ") 
  for res in mt: 
    print(res)
```

### Задача 2

Нужно вывести на печать матрицу размером N*N, заполненную числами натурального ряда, начиная с первого элемента и двигаясь по спирали:

---

В зависимости от введенных данных результат может выглядеть так:

image.png

```python
# Ваше решение
n = int(input())
arr = [[0 for i in range(n)] for j in range(n)]
dx, dy = 1, 0
x, y = 0, 0
for i in range(1, n**2+1):
    arr[x][y] = i
    nx, ny = x+dx, y+dy
    if 0 <= nx < n and 0 <= ny < n and arr[nx][ny] == 0:
        x, y = nx, ny
    else:
        dx, dy = -dy, dx
        x, y = x+dx, y+dy

for i in range(n):
    for j in range(n):
        print(arr[j][i], end=" ")
    print()
```

```python
# Ваше решение
import math
import random

no_of_balls = 100
x = [random.triangular() for i in range(no_of_balls)]
y = [random.gauss(0.5, 0.25) for i in range(no_of_balls)]
colors = [random.randint(1, 4) for i in range(no_of_balls)]
areas = [math.pi * random.randint(5, 15)**2 for i in range(no_of_balls)]
# draw the plot
plt.figure()
plt.scatter(x, y, s=areas, c=colors, alpha=0.7)
plt.axis([0.0, 1.0, 0.0, 1.0])
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
